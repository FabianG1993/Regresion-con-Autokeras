```python
import autokeras as ak
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

Aplicando AutoKeras al boston dataset

```python
# carga de datos
boston = pd.read_csv('/content/drive/MyDrive/AutoML/boston (1).csv')
X_boston = boston.drop('medv', axis=1)
y_boston = boston['medv']
```

```python
boston.info()
```

```python
# establecer datos para entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_boston, y_boston, random_state=42)
```

```python
# crear regresor con la clase StructuredDataRegressor 
regresor = ak.StructuredDataRegressor(overwrite=True, max_trials=3)

# ajustar el modelo a los datos
regresor.fit(X_train, y_train)

# hacer predicciones
y_pred = regresor.predict(X_test)
```

```python
# evaluar modelo con MSE y RMSE
y_pred = y_pred.ravel() # version unidimensional de y_pred

#calculo MSE
mse = np.mean((y_test - y_pred) ** 2)
print(f"Mean squared error: {mse:.2f}")

#calculo RMSE
rmse = np.sqrt(mse)
print(f"Root mean squared error: {rmse:.2f}")
```

En el conjunto de datos de Boston, donde el valor medio de la vivienda (medv) se mide en miles de dólares, un error cuadrático medio (RMSE) de 3,47 se considera bueno.
Es importante tener en cuenta que el valor medio de la vivienda en el conjunto de datos de Boston oscila entre aproximadamente 5000-50000, lo que significa que la variable objetivo tiene un rango de valores relativamente pequeño. Por lo tanto, el RMSE de 3,47 representa una parte significativa del rango de valores de medv.
